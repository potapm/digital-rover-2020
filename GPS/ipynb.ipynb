```python
import pandas as pd
import numpy as np


LEP=pd.read_csv('ЛЭП.csv',header=None,sep=';')
LEP
```

```python
import pandas as pd
from folium.plugins import HeatMap
from folium.plugins import FastMarkerCluster

import folium
#Создание карты
hmap = folium.Map(location=[52.692036, 39.627396], zoom_start=16)
for i in LEP.index: 
    folium.CircleMarker([LEP[1].loc[i], LEP[2].loc[i]],radius=1, color='blue').add_to(hmap)

#Сохраняем полученную карту в html файл
hmap.save('REZ3.html')
```
